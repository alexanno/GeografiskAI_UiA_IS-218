```python
import os
import pandas as pd

# Define the path to the CSV file
csv_file_path = r'c:/dev/GeografiskAI_UiA_IS-218/tmp.csv'

# Read the CSV file using pandas
df = pd.read_csv(csv_file_path, delimiter='\t')

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Create folder name based on "uke" and "modul"
    folder_name = f"{row['Uke']}-{row['Modul']}"
    os.makedirs(folder_name, exist_ok=True)
    
    # Create README.md file inside the folder
    readme_path = os.path.join(folder_name, 'README.md')
    with open(readme_path, 'w', encoding='utf-8') as readme_file:
        # Write the corresponding row as text in the README.md
        print(row)
        readme_file.write(row.to_string())
```
