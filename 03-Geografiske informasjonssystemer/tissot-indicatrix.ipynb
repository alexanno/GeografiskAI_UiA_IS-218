# Tissot indicatrix - visualisering av feil i projeksjoner
Tissot's indicatrix is a method to visualize the distortion caused by map projections. In this notebook, we will explore various projections available in the `cartopy` library and visualize the distortion using Tissot's indicatrix. The projections we will explore include:

- PlateCarree
- Mollweide
- Robinson
- InterruptedGoodeHomolosine
- NorthPolarStereo

We will use the `tissotindicatrix` function to generate the visualizations.

### Generate maps with Tissot Indicatrix for several projections

Cartopy is a Python library designed for geospatial data processing to produce maps and other geospatial data analyses. It provides tools for creating maps with various projections and adding features like coastlines, borders, and gridlines.

- **Mollweide Projection**
  - `tissotindicatrix(ccrs.Mollweide())`
  - An equal-area projection that represents the entire Earth in an elliptical shape. It is often used for global maps to show distributions and densities.

- **Robinson Projection**
  - `tissotindicatrix(ccrs.Robinson())`
  - A compromise projection that balances size and shape distortions, making it visually appealing for world maps. It is neither equal-area nor conformal.

- **Interrupted Goode Homolosine Projection**
  - `tissotindicatrix(ccrs.InterruptedGoodeHomolosine())`
  - An equal-area, composite projection that minimizes distortion by interrupting the map in the oceans. It is useful for thematic maps of landmasses.

- **North Polar Stereographic Projection**
  - `tissotindicatrix(ccrs.NorthPolarStereo())`
  - A conformal projection that shows the area around the North Pole. It preserves angles, making it useful for polar navigation and meteorological charts.

- **Plate Carree Projection**
  - `tissotindicatrix(ccrs.PlateCarree())`
  - A simple cylindrical projection where the meridians and parallels are straight and perpendicular. It is often used for global maps and is the default projection in many mapping libraries.

These projections are used in Cartopy to visualize geospatial data with different emphasis on area, shape, and visual appeal.

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

import matplotlib.pyplot as plt

import cartopy.crs as ccrs

def tissotindicatrix(proj=ccrs.PlateCarree()):
    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(1, 1, 1, projection=proj)

    # make the map global rather than have it zoom in to
    # the extents of any plotted data
    ax.set_global()

    ax.stock_img()
    ax.coastlines()

    ax.tissot(facecolor='orange', alpha=0.9)

    plt.show()

tissotindicatrix()
tissotindicatrix(ccrs.Mollweide())
tissotindicatrix(ccrs.Robinson())
tissotindicatrix(ccrs.InterruptedGoodeHomolosine())
tissotindicatrix(ccrs.NorthPolarStereo())
```
